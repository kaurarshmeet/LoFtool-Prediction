```python
#read stuff in 
import numpy as np 
import pandas as pd 
df_loftool = pd.read_csv('df_loftool.csv')  
```

```python
#Univariate Feature Selection (F Regression)

#got input from chatgpt here
#learned from here too: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html
#learned (also) from here: https://stephenallwright.com/cross_val_score-sklearn
#https://grsahagian.medium.com/what-is-random-state-42-d803402ee76b 

import sklearn
from sklearn.neighbors import KNeighborsRegressor
import numpy as np 
import pandas as pd 

KNN = KNeighborsRegressor()

X = df_loftool[['POS', 'cDNA_position', 'CDS_position', 'Protein_position', 'STRAND', 'CHROM', 'SYMBOL', 'Feature', 'EXON', 'Codons']]
y = df_loftool['LoFtool'] 

#regular evaluation 
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.3, random_state = 41)  
KNN.fit(X_train, y_train) 

y_pred = KNN.predict(X_test)
r2 = sklearn.metrics.r2_score(y_test,y_pred) 
mse = sklearn.metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse) 
mae = sklearn.metrics.mean_absolute_error(y_test, y_pred) 
ev = sklearn.metrics.explained_variance_score(y_test, y_pred) 
print("r2:", r2, "mse:", mse, "rmse:", rmse, "mae:", mae, "ev:", ev) 

y_pred_training = KNN.predict(X_train) 
r2_training = sklearn.metrics.r2_score(y_train, y_pred_training) 
mse_training = sklearn.metrics.mean_squared_error(y_train, y_pred_training)
rmse_training = np.sqrt(mse_training)
mae_training = sklearn.metrics.mean_absolute_error(y_train, y_pred_training)
ev_training = sklearn.metrics.explained_variance_score(y_train, y_pred_training)
print("r2_train:", r2_training, "mse_train:", mse_training, "rmse_training:", rmse_training, "mae_training:", mae_training, "ev_training:", ev_training) 

#cross validation 
cross_validation = sklearn.model_selection.cross_val_score(KNN, X, y, cv = 5)
sklearn.metrics.get_scorer_names()
print("Cross-validation score:", cross_validation)
print("Average cross-validation score:", cross_validation.mean()) 

r2 = sklearn.model_selection.cross_val_score(KNN, X, y, cv = 5, scoring = 'r2')
print("R2 Score:", r2) 
print("Average R2 score:", r2.mean()) 

mse = sklearn.model_selection.cross_val_score(KNN, X, y, cv=5, scoring = 'neg_mean_squared_error') 
print("MSE score:", mse) 
print("Average MSE score:", mse.mean()) 

rmse = sklearn.model_selection.cross_val_score(KNN, X, y, cv=5, scoring = 'neg_root_mean_squared_error') 
print("RMSE score:", rmse)
print("Average RMSE score:", rmse.mean()) 

mae = sklearn.model_selection.cross_val_score(KNN, X, y, cv=5, scoring = 'neg_mean_absolute_error') 
print("MAE score:", mae) 
print("Average MAE score:", mae.mean()) 

explained_variance = sklearn.model_selection.cross_val_score(KNN, X, y, cv = 5, scoring = 'explained_variance') 
print("Explained Variance:", explained_variance) 
print("Average Explained Variance:", explained_variance.mean()) 
```
